```python
def plot_stationarity(x,f,h):

    # Distances between points
    dx = x[1:] - x[:-1]
    if dx.max() - dx.min() > 1e2 or np.any(dx<0):
        print('x needs to be increasing and evenly spaced')
        return
    dx = np.cumsum(np.concatenate([np.array([0]),dx]))
    h_idx = np.argmin(np.abs(dx - h[:,np.newaxis]),axis=1) #Closest multiple of dx to h
    
    # Covariances at offset h
    C = np.cov(f.T)
    Ch = [np.diagonal(C,offset=h) for h in h_idx]

    pad = [(int(np.floor((x.size-row.size)/2)),int(np.ceil((x.size-row.size)/2))) for row in Ch]
    Ch_mat = np.array([\
        np.concatenate([np.full(p[0],np.nan), row,np.full(p[1],np.nan)]) \
        for (p,row) in zip(pad,Ch)])

    # Figure
    fig, ax = plt.subplots()
    ax.plot(x, Ch_mat.T,'o-')
    ax.legend(['k=%.1f' % h for h in dx[h_idx]])

    ax.set_xlabel(r'$x$')

    return fig, ax
```

## Fixed uniform intensity

```python
intensity = 10
s2_0 = 5
T = [-1,1]
prior_b_sig2 = 1.0
prior_w_sig2 = 1.0 # This is \tilde{sigma}^2_w, not \tilde{sigma}^2_w. The network uses \tilde{sigma}^2_w 

x_plot = np.linspace(-2,2,1000)
```

#### Analytical

```python
def cov_approx(x1, x2, prior_b_sig2, prior_w_sig2, s2_0, intensity):
    return prior_b_sig2 + prior_w_sig2*np.exp(-s2_0*intensity**2*((x1-x2)/2)**2)

def cov(x1, x2, prior_b_sig2, prior_w_sig2, s2_0, intensity, C0, C1):
    s2 = s2_0*intensity**2
    xm = (x1+x2)/2
    
    C0_tilde = (C0-xm)*np.sqrt(2*s2)
    C1_tilde = (C1-xm)*np.sqrt(2*s2)
    
    return prior_b_sig2 + prior_w_sig2*np.exp(-s2*((x1-x2)/2)**2) \
        *(norm.cdf(C1_tilde)-norm.cdf(C0_tilde))
```

```python
h = .1
cov_plot = cov(x_plot-h, x_plot+h, prior_b_sig2=prior_b_sig2, prior_w_sig2=prior_w_sig2, \
                s2_0=s2_0, intensity=intensity, C0=T[0], C1=T[1])

plt.plot(x_plot, cov_plot)
```

## Uniform intensity with Gamma prior

```python
intensity = 5
s2_0 = 5
T = [-1,1]
prior_b_sig2 = 1.0
prior_w_sig2 = 1.0

x_plot = np.linspace(-2,2,1000)
```

#### Analytical

```python
# After Lambda fix
def V(x1, x2, s, C0, C1, alpha, beta):
    dx2 = (1/4)*s**2*(x1-x2)**2 + beta
    xm = (x1+x2)/2
    Ct0 = xm - C0
    Ct1 = C1 - xm
    return dx2**(-alpha) * \
            (Ct0*hyp2f1(1/2,alpha,3/2, -s**2*Ct0**2/dx2) + \
             Ct1*hyp2f1(1/2,alpha,3/2, -s**2*Ct1**2/dx2))

def Cov(x1, x2, s, C0, C1, alpha, beta, sig2w, sig2b):
    E_K = intensity*(C1-C0)
    print(E_K)
    return sig2b + sig2w\
            *E_K * 1/(C1-C0)*beta**(alpha)\
            *V(x1, x2, s, C0, C1, alpha, beta)
```

```python
Cov(0, 0, \
    s=np.sqrt(s2_0), C0=T[0], C1=T[1], \
    alpha=intensity**2, \
    beta=1.0, \
    sig2w=prior_w_sig2, \
    sig2b=prior_b_sig2)
```

```python
def cov_porbnet_gamma_intensity(x1, x2, C0, C1, b_sig2, w_sig2, s2_0, intensity_alpha, intensity_beta):
    '''
    Covariance for porbnet with 1d inputs and uniform intensity with gamma prior
    '''
    dx2 = (1/4)*s2_0*(x1-x2)**2 + intensity_beta
    xm = (x1+x2)/2
    Ct0 = xm - C0
    Ct1 = C1 - xm
    V = dx2**(-intensity_alpha) * \
            (Ct0*hyp2f1(1/2,intensity_alpha,3/2, -s2_0*Ct0**2/dx2) + \
             Ct1*hyp2f1(1/2,intensity_alpha,3/2, -s2_0*Ct1**2/dx2))

    E_K = mean_sqrt_gamma(intensity_alpha, intensity_beta)
    print(E_K)
    return b_sig2 + w_sig2*E_K*intensity_beta**(intensity_alpha)*V

def mean_sqrt_gamma(alpha, beta):
    try:
        return np.exp(loggamma(alpha+0.5)-loggamma(alpha)-0.5*np.log(beta))
    except:
        # in case of numerical issue, use approximation
        return alpha**(0.5)*beta**(-0.5)
```
