```python
def plot_stationarity(x,f,h,ax=None):

    # Distances between points
    dx = x[1:] - x[:-1]
    if dx.max() - dx.min() > 1e2 or np.any(dx<0):
        print('x needs to be increasing and evenly spaced')
        return
    dx = np.cumsum(np.concatenate([np.array([0]),dx]))
    h_idx = np.argmin(np.abs(dx - h[:,np.newaxis]),axis=1) #Closest multiple of dx to h
    
    # Covariances at offset h
    C = np.cov(f.T)
    Ch = [np.diagonal(C,offset=h) for h in h_idx]

    pad = [(int(np.floor((x.size-row.size)/2)),int(np.ceil((x.size-row.size)/2))) for row in Ch]
    Ch_mat = np.array([\
        np.concatenate([np.full(p[0],np.nan), row,np.full(p[1],np.nan)]) \
        for (p,row) in zip(pad,Ch)])

    # Figure
    if ax is None:
        fig, ax = plt.subplots()
    ax.plot(x, Ch_mat.T,'o-',alpha=.025)
    ax.legend(['t=%.1f' % h for h in dx[h_idx]], loc='upper right')

    ax.set_xlabel(r'$x$')
    
    return ax
```

## Fixed uniform intensity

```python
intensity = 10
s2_0 = 4
T = [-1,1]
prior_b_sig2 = 1e-16
prior_w_sig2 = 1.0

x_plot = np.linspace(-1.5,1.5,1000)
```

```python
intensity_func = util_porbnet.Piecewise(np.array([T[0],T[1]]),np.array([intensity])) 

dim_hidden_initial = (T[1]-T[0])*intensity

net = networks_porbnet.RBFN(dim_in=1, dim_hidden_initial=dim_hidden_initial, \
                            dim_hidden_max=3*dim_hidden_initial, \
                            dim_out=1, intensity=intensity_func, s2_0=s2_0, \
                            prior_w_sig2 = prior_w_sig2*np.sqrt(np.pi/s2_0), prior_b_sig2 = prior_b_sig2, \
                            sig2 = .01,\
                            prior_intensity_alpha = intensity, prior_intensity_beta = 1.0)

torch.manual_seed(0)
f_samp = net.sample_functions_prior(torch.from_numpy(x_plot).reshape(-1,1), n_samp=1000, sample_K=True, sample_intensity=False).detach().numpy()

```

```python
h_list = np.array([0,.1,.2])
fig, ax = plt.subplots(figsize=(5,3))
colors = ['C0','C1','C2']

for i, h in enumerate(h_list): 
    cov_plot = util_porbnet.cov_porbnet_fixed_intensity(x1=x_plot-h/2, \
                                             x2=x_plot+h/2, \
                                             C0=T[0], \
                                             C1=T[1], \
                                             b_sig2=prior_b_sig2, \
                                             w_sig2=prior_w_sig2, \
                                             s2_0=s2_0, \
                                             intensity=intensity)
    ax.plot(x_plot, cov_plot, color=colors[i])
    
plot_stationarity(x_plot,f_samp,h=h_list,ax=ax)
ax.set_ylim(-.1,2.1)
fig.tight_layout()
```

## Uniform intensity with Gamma prior

```python
intensity = 10
s2_0 = 1
T = [-1,1]
prior_b_sig2 = 1e-16
prior_w_sig2 = 1.0

x_plot = np.linspace(-1.5,1.5,1000)
```

```python
alpha = util_porbnet.alpha_for_sqrt_gamma(beta=1.0, K=intensity)
```

```python
intensity_func = util_porbnet.Piecewise(np.array([T[0],T[1]]),np.array([intensity])) 

dim_hidden_initial = (T[1]-T[0])*intensity

net = networks_porbnet.RBFN(dim_in=1, dim_hidden_initial=dim_hidden_initial, \
                            dim_hidden_max=3*dim_hidden_initial, \
                            dim_out=1, intensity=intensity_func, s2_0=s2_0, \
                            prior_w_sig2 = prior_w_sig2*np.sqrt(np.pi/s2_0), prior_b_sig2 = prior_b_sig2, \
                            sig2 = .01,\
                            prior_intensity_alpha = alpha, prior_intensity_beta = 1.0)

torch.manual_seed(0)
f_samp = net.sample_functions_prior(torch.from_numpy(x_plot).reshape(-1,1), n_samp=1000, sample_K=True, sample_intensity=True).detach().numpy()

```

```python
def cov_porbnet_gamma_intensity(x1, x2, C0, C1, b_sig2, w_sig2, s2_0, intensity_alpha, intensity_beta):
    '''
    Covariance for porbnet with 1d inputs and uniform intensity with gamma prior
    '''
    dx2 = (1/4)*s2_0*(x1-x2)**2 + intensity_beta
    xm = (x1+x2)/2
    Ct0 = xm - C0
    Ct1 = C1 - xm
    V = dx2**(-intensity_alpha) * \
            (Ct0*hyp2f1(1/2,intensity_alpha,3/2, -s2_0*Ct0**2/dx2) + \
             Ct1*hyp2f1(1/2,intensity_alpha,3/2, -s2_0*Ct1**2/dx2))

    E_lambda = mean_sqrt_gamma(intensity_alpha, intensity_beta)

    cov =  b_sig2 + w_sig2*E_lambda*intensity_beta**(intensity_alpha)*V
    return cov, E_lambda, V
```

```python
from scipy.special import hyp2f1, gamma
```

```python
def mean_sqrt_gamma(alpha, beta):
    try:
        return np.exp(loggamma(alpha+0.5)-loggamma(alpha)-0.5*np.log(beta))
    except:
        # in case of numerical issue, use approximation
        return alpha**(0.5)*beta**(-0.5)
```

```python
intensity = 4
s2_0 = 1
T = [-1,1]
prior_b_sig2 = 1e-16
prior_w_sig2 = 1.0

alpha = util_porbnet.alpha_for_sqrt_gamma(beta=1.0, K=intensity)

cov_porbnet_gamma_intensity(x1=0,\
                           x2=0,\
                           C0=T[0],\
                           C1=T[1],\
                           b_sig2=prior_b_sig2,\
                           w_sig2=prior_w_sig2,\
                           s2_0=s2_0,\
                           intensity_alpha=alpha,\
                           intensity_beta=1.0)
```

```python
intx = np.linspace(.1,20,30)
cov = np.zeros(len(intx))
E_lambda = np.zeros(len(intx))
V = np.zeros(len(intx))

for i in range(len(intx)):
    alpha = util_porbnet.alpha_for_sqrt_gamma(beta=1.0, K=intx[i])

    cov[i], E_lambda[i], V[i] = cov_porbnet_gamma_intensity(x1=0,\
                               x2=0,\
                               C0=T[0],\
                               C1=T[1],\
                               b_sig2=prior_b_sig2,\
                               w_sig2=prior_w_sig2,\
                               s2_0=s2_0,\
                               intensity_alpha=alpha,\
                               intensity_beta=1.0)
```

```python
from scipy.stats import norm
```

```python
def V0(x1, x2, C0, C1, s2_0, intensity):
    s2 = s2_0*intensity**2
    xm = (x1+x2)/2
    
    C0_tilde = (C0-xm)*np.sqrt(2*s2)
    C1_tilde = (C1-xm)*np.sqrt(2*s2)

    return (norm.cdf(C1_tilde)-norm.cdf(C0_tilde))

V0eval = lambda intensity: V0(0,0,T[0],T[1],s2_0,intensity)
```

```python
plt.plot(intx, V0eval(intx), '-o')
```

So in gamma case or non-gamma case, the "modulation term" is impacted by the intensity (either directly in non-gamma case or through alpha parameter in gamma case). When intensity increases, modulation term approaches a constant. But without making the infinite C argument, the variance does depend on intensity for small values
